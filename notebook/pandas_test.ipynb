```python
import requests
irish_data = requests.get('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data').text
```

```python
from StringIO import StringIO
```

```python
iris = pd.read_csv(StringIO(irish_data), names=['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Name'])
```

```python
iris.head()
```

```python
iris.assign(sepal_ratio = iris['SepalWidth'] / iris['SepalLength']).head()
```

```python
iris.assign(sepal_ratio = lambda x: (x['SepalWidth'] / x['SepalLength'])).plot()
```

```python
iris.query('SepalLength > 5')\
.assign(SepalRatio = lambda x: x.SepalWidth / x.SepalLength,
PetalRatio = lambda x: x.PetalWidth / x.PetalLength)\
.plot(kind='scatter', x='SepalRatio', y='PetalRatio')
```
